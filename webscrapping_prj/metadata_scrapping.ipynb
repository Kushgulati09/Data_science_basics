```python
from PIL import Image
from PIL.ExifTags import TAGS

import os
import glob

import csv

import pandas as pd

from sqlalchemy import create_engine, engine

# This function will write the metadata fetched into a csv file
def write(data_ , file_name):
    with open(file_name , 'a' , encoding = 'UTF8' , newline = '') as f:
        writer = csv.writer(f)
        writer.writerow(data_)
        f.close()

# These are the fields we are interested to fetch
header = ["Filename","Image Size",
        "Image Height",
        "Image Width",
        "Image Format",
        "Image Mode",
        "Image is Animated",
        "Frames in Image"]

# Open a new file to write
with open('Images_.csv' , 'w' , encoding = 'UTF8' , newline = '') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    f.close()

files = []
path = "D:\\KUSH GULATI\\dataset\\mouse\\*.jpg"

for file in glob.iglob(path, recursive=True):
    files.append(file)


for i in files:
    imagename = i

    
    image = Image.open(imagename)

    
    info_dict = {
        "Filename": image.filename,
        "Image Size": image.size,
        "Image Height": image.height,
        "Image Width": image.width,
        "Image Format": image.format,
        "Image Mode": image.mode,
        "Image is Animated": getattr(image, "is_animated", False),
        "Frames in Image": getattr(image, "n_frames", 1)
    }

    data_ = list(info_dict.values())
    write(data_,'Images_.csv')

df = pd.read_csv("Images_.csv")

engine = create_engine('sqlite:///:memory:')
df.to_sql('Images', engine)

sqldf = pd.read_sql('Images', engine, index_col = 'index')
sqldf
```
