```python
import os
import glob
import pandas as pd

k = []

path = "D:\\KUSH GULATI\\dataset\\mouse\\*.jpg"


for files in glob.iglob(path, recursive=True):
    print(files)
    k.append(files)
    
```

```python
df = pd.DataFrame({'Path': k})
```

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite:///:memory:') #creates a temporary engine database that run in the memory
df.to_sql('Images', engine) #store the dataframe in the engine

sqldf = pd.read_sql('Images', engine, index_col='index') #extract databse from the engine
sqldf 
```
